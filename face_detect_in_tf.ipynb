```python
import cv2
# Get user supplied values
#imagePath = '/Users/maheshworpaudel/Desktop/Projects/test1.jpeg'
#imagePath = '/Users/maheshworpaudel/Downloads/IMG_5689.JPG'
#imagePath = '/Users/maheshworpaudel/Downloads/IMG_5641.png'
imagePath = '/Users/maheshworpaudel/Downloads/IMG_5652-2.png'
cascPath = "haarcascade_frontalface_default.xml"
# Create the haar cascade
faceCascade = cv2.CascadeClassifier(cascPath)
# Read the image
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=10,
    minSize=(250, 250),
    #flags = cv2.CV_16S.CV_HAAR_SCALE_IMAGE
)

print("Found {0} faces!".format(len(faces)))
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 10))
plt.imshow(image)
from matplotlib.patches import Rectangle
for (x, y, w, h) in faces:
    plt.gca().add_patch(Rectangle((x,y),w,h,
                        edgecolor='red',
                        facecolor='none',
                        lw=1))
plt.show()
```
